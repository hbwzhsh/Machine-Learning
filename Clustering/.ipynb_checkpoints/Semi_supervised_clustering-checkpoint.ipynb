## Semi-supervised clustering algorithm

This model is similar to the k-means clustering algorithm with the exception that the cluster centroids are based on the labeled examples, thus the cluster will not be updated. This comes with the assumption that there are labeled examples for every class. Based on these cluster centroids, unlabeled data will be assigned to the closest centroid based on squared distance. Given this is a generative model, there is the obvious constraint that there must be enough labeled data for each class to learn a valid distribution of data per class.

Also, given that this is a simple generative model, I forsee that there might be issues in how outliers in the labeled data per class might skew the distribution away from something valid. Thus, I also experiment with "forgetting" the outliers for any given class when trying to find the cluster centroids. I will be analyzing this algorithm in terms of accuracy and testing whether "forgetting" outliers has an impact on model performance.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
import latex
import random
from sklearn.metrics import accuracy_score

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
breast_cancer = datasets.load_breast_cancer()
X = breast_cancer.data[:,:2]  # 2x features
y = breast_cancer.target
print(X.shape)
print(y.shape)
```

### Visualizing the data

```python
# getting the data corresponding with cancer and no cancer
X_no = X[y == 0]
X_yes = X[y == 1]
print("Number of no cancer ex.:",X_no.shape)
print("Number of yes cancer ex.:",X_yes.shape)
# getting subset of data to act as limited labeled examples
X_no_lim = X_no[:40]
X_yes_lim = X_yes[:40]
X_no_unlim = X_no[40:]
X_yes_unlim = X_yes[40:]
```

```python
# visualizing the example data
# data seems to be seperable, but there is definitely some overlap
plt.scatter(X[:,0],X[:,1],c=y,cmap="magma",alpha=0.2)
plt.scatter(X_no_lim[:,0],X_no_lim[:,1],c="blue",label="no cancer sample")
plt.scatter(X_yes_lim[:,0],X_yes_lim[:,1],c="orange",label="cancer sample")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### Visualizing the model without outliers being removed

```python
# returns the centroid for a given set of 2D data
def get_centroid(data):
    cent_x = np.average(data[:,0])
    cent_y = np.average(data[:,1])
    ret_arr = np.array([cent_x,cent_y])
    ret_arr.shape = (1,2)
    return ret_arr
```

```python
# visualizing the centroid which will be created for each of the samples
centroid_no = get_centroid(X_no_lim)
centroid_yes = get_centroid(X_yes_lim)

plt.scatter(X[:,0],X[:,1],c=y,cmap="magma",alpha=0.2)
plt.scatter(centroid_no[:,0],centroid_no[:,1],c="blue",label="no cancer centroid")
plt.scatter(centroid_yes[:,0],centroid_yes[:,1],c="orange",label="cancer centroid")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Calculate the squared distance between centroid and example
def squared_distance(example, centroid):
    dist = np.sum((example - centroid)**2)
    return np.asscalar(dist) #shape = (1)
```

```python
# Assignes data points to closest centroid and returns values of data for each centroid
def cluster_assignment(X,cents):
    preds = [] # holds 0,1 predictions
    cent_one = []
    cent_two = []
    for i in range(X.shape[0]):
        dist_cent_one = squared_distance(X[i,:], cents[0])
        dist_cent_two = squared_distance(X[i,:], cents[1])
        if dist_cent_one < dist_cent_two:
            cent_one.append(X[i,:]) # closer to centroid one
            preds.append(0)
        else:
            cent_two.append(X[i,:]) # closer to centroid two
            preds.append(1)
            
    return cent_one,cent_two,preds
```

```python
# Getting predictions
cents = [centroid_no,centroid_yes]
X_unlim = np.concatenate((X_no_unlim,X_yes_unlim)) # all unlabeled data
y_act = [0 for i in range(len(X_no_unlim))] + [1 for i in range(len(X_yes_unlim))] # correct labels
cent_one,cent_two,preds = cluster_assignment(X_unlim,cents)
```

```python
# Reshaping predictions to arrays
cent_one = np.array(cent_one)
cent_two = np.array(cent_two)
```

```python
print("accuracy score of standard clustering:",accuracy_score(y_act,preds))

plt.scatter(cent_one[:,0],cent_one[:,1],c="blue",label="no cancer preds")
plt.scatter(cent_two[:,0],cent_two[:,1],c="orange",label="cancer preds")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Class Predictions")
plt.legend()
plt.show()
```

### Visualizing the model with outliers being removed

```python
# returns the average distance between data and a centroid
def get_avg_distance(centroid,data):
    all_distances = []
    for i in range(data.shape[0]):
        distance = squared_distance(data[i,:], centroid)
        all_distances.append(distance)
        
    return np.average(all_distances)
```

```python
# outliers are defined as a distance from the centroid 2x greater than the average distance
# this returns the data examples defined as outlier and not outlier
def get_outliers(centroid,data,avg_dist):
    outliers = []
    not_outliers = []
    for i in range(data.shape[0]):
        distance = squared_distance(data[i,:], centroid)
        if distance > (2 * avg_dist):
            outliers.append(data[i,:])
        else:
            not_outliers.append(data[i,:])
            
    return np.array(outliers),np.array(not_outliers)
```

```python
avg_dist = get_avg_distance(centroid_no,X_no_lim)
outliers_nocancer,not_outliers_nocancer = get_outliers(centroid_no,X_no_lim,avg_dist)
print(avg_dist) # average distance example for the no cancer labled subset
```

```python
# getting the new centroid from the data which removes the outliers
centroid_no_new = get_centroid(not_outliers_nocancer)
```

```python
# visualizing which points would include the outliers in this case (for no cancer data)
# also prints out the old centroid and the new centroid

plt.scatter(X[:,0],X[:,1],c=y,cmap="magma",alpha=0.2)
plt.scatter(outliers_nocancer[:,0],outliers_nocancer[:,1],c="orange",label="outliers")
plt.scatter(not_outliers_nocancer[:,0],not_outliers_nocancer[:,1],c="blue",label="not outliers")
plt.scatter(centroid_no[:,0],centroid_no[:,1],c="yellow",label="old centroid")
plt.scatter(centroid_no_new[:,0],centroid_no_new[:,1],c="red",label="new centroid")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Visualizing effect of outliers for no cancer data")
plt.legend()
plt.show()
```

```python
# Doing same outlier analysis for the yes cancer data
centroid_yes = get_centroid(X_yes_lim)
avg_dist_yes = get_avg_distance(centroid_yes,X_yes_lim)
outliers_cancer,not_outliers_cancer = get_outliers(centroid_yes,X_yes_lim,avg_dist_yes)
centroid_yes_new = get_centroid(not_outliers_cancer)

print(avg_dist_yes) # the average distance for the cancer data is smaller
```

```python
# visualizing which points would include the outliers in this case (for yes cancer data)

plt.scatter(X[:,0],X[:,1],c=y,cmap="magma",alpha=0.2)
plt.scatter(outliers_cancer[:,0],outliers_cancer[:,1],c="orange",label="outliers")
plt.scatter(not_outliers_cancer[:,0],not_outliers_cancer[:,1],c="blue",label="not outliers")
plt.scatter(centroid_yes[:,0],centroid_yes[:,1],c="black",label="old centroid")
plt.scatter(centroid_yes_new[:,0],centroid_yes_new[:,1],c="red",label="new centroid")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Visualizing effect of outliers for yes cancer data")
plt.legend()
plt.show()
```

#### Ultimately the cluster centroid change slightly but to no significant degree.

```python
# using the new centroid centers based on "forgetting" outliers to test model accuracy

cents = [centroid_no_new,centroid_yes_new] # using updated centroid centers
X_unlim = np.concatenate((X_no_unlim,X_yes_unlim)) # all unlabeled data
y_act = [0 for i in range(len(X_no_unlim))] + [1 for i in range(len(X_yes_unlim))] # correct labels
cent_one,cent_two,preds = cluster_assignment(X_unlim,cents)

# Reshaping predictions to arrays
cent_one = np.array(cent_one)
cent_two = np.array(cent_two)
```

```python
print("accuracy score of standard clustering:",accuracy_score(y_act,preds))

plt.scatter(cent_one[:,0],cent_one[:,1],c="blue",label="no cancer preds")
plt.scatter(cent_two[:,0],cent_two[:,1],c="orange",label="cancer preds")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Class Predictions")
plt.legend()
plt.show()
```

#### Ultimately for this distribution of data, "forgetting" the outliers leads to a slightly smaller accuracy. However, overall it seems that this semi-supervised architecture does okay, at least when classes are reasonably separable.
