## Cleaning Image Data
I will be resizing the images, and then labeling the data such that it meets YOLO specifications - specifically that the output will be a matrix of shape 13x13x(5+#animal_types) = 13x13x7. The output shape represents the ground truth bounding box data for each cell in the image. Each of these cells will be 32x32 pixels.

Dataset used: The Oxford-IIIT Pet Dataset

Model I referenced - https://github.com/llSourcell/YOLO_Object_Detection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import latex
import os
import xmltodict
from PIL import Image
import matplotlib.patches as patches
from skimage.transform import resize
import math

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# stores all the data about each image
train_dict = {}
```

```python
"""
Getting the name, species, images, and bounding box data
Note - not all data has bounding boxes, so we are ignoring those
species: 1 (Cat), 2 (Dog)
filename: the text file that contains info about a subsect of images
"""
def get_image_data(filename):
    animal_dict = {}
    with open("../../data/yolodata/annotations/"+filename) as file:
        for line in file.readlines():
            split = line.split(" ")
            name = split[0]
            species = split[2]
            animal_dict[name] = {"species":species,"xmin":0,"xmax":0,"ymin":0,"ymax":0,"width":0, "length":0,"img":0}
            animal_dict = get_xml_data(name,animal_dict)
            if animal_dict[name]["xmin"] == 0 and animal_dict[name]["xmax"] == 0:
                del animal_dict[name] # there was no bounding box data

        return animal_dict
```

```python
"""
Gets the bounding box data - xmin,ymin,xmax,ymax, along with image size
"""
def get_xml_data(name, animal_dict):
    try: # If fail then no xml
        xml_name = name + ".xml"
        file = open("../../data/yolodata/annotations/xmls/" + xml_name) # Check if xml exists
        doc = doc = xmltodict.parse(file.read())
        animal_dict[name]['xmin'] = int(doc['annotation']['object']['bndbox']['xmin'])
        animal_dict[name]['xmax'] = int(doc['annotation']['object']['bndbox']['xmax'])
        animal_dict[name]['ymax'] = int(doc['annotation']['object']['bndbox']['ymax'])
        animal_dict[name]['ymin'] = int(doc['annotation']['object']['bndbox']['ymin'])
        animal_dict[name]['width'] = int(doc['annotation']['size']['width'])
        animal_dict[name]['length'] = int(doc['annotation']['size']['height'])
        file.close()    
        return animal_dict
    except:
        return animal_dict
```

```python
train_dict = get_image_data("trainval.txt")
len(train_dict.keys()) # The number of items in training set (note-no other data was labeled)
```

```python
"""
Getting the image data for each training data
folder: specifies where the image data is held
"""
def get_image_data(train_dict):
    for animal in train_dict.keys():
        img_name = animal+".jpg"
        try:
            file = Image.open("../../data/yolodata/images/"+img_name)
            img = np.array(file)
            train_dict[animal]['img'] = img # saving np array of image
            file.close()
        except:
            continue
    return train_dict
```

```python
# This will store all my training data including augmented data
aug_train = {}
```

```python
"""
First, I am going to do some data augmentation, namely flipping the image along the horizontal axis 
This will allow me to increase my data set
Note - I am using the data dictionary with small images removed
"""
for animal in ftrain_dict.keys():
    xmin = train_dict[animal]['xmin']
    xmax = train_dict[animal]['xmax']
    ymin = train_dict[animal]['ymin']
    ymax = train_dict[animal]['ymax']
    width = train_dict[animal]['width']
    length = train_dict[animal]['length']
    species = train_dict[animal]['species']
    img = train_dict[animal]['img']
    aug_train[animal] = {'species':species,'xmin':xmin,'xmax':xmax,'ymin':ymin,'ymax':ymax,'width':width,'length':length,'img':img}
    # update image
    img = np.flip(img,axis=1) # Rotation about horizontal axis
    # update xmin and xmax
    box_width = xmax - xmin
    xmax = width - xmin
    xmin = xmax - box_width
    # Augmented data denoted w/ tailing _2
    aug_train[animal+'_2'] = {'species':species,'xmin':xmin,'xmax':xmax,'ymin':ymin,'ymax':ymax,'width':width,'length':length,'img':img}
```

```python
# This is an example of flipping
xmin = train_dict['Abyssinian_101']['xmin']
ymin = train_dict['Abyssinian_101']['ymin']
width = train_dict['Abyssinian_101']['xmax'] - xmin
length = train_dict['Abyssinian_101']['ymax'] - ymin
imgwidth = train_dict['Abyssinian_101']['width']

new_xmax = imgwidth - xmin
print(new_xmax)
new_xmin = new_xmax - width
print(new_xmin)

fig,ax = plt.subplots(1)
ax.imshow(np.flip(train_dict['Abyssinian_101']['img'],axis=1))
rect = patches.Rectangle((new_xmin,ymin),width,length,linewidth=1,edgecolor='r',facecolor='none')
ax.add_patch(rect)

plt.show()
```

#### I am resizing all of the images to 416x416 using image spline interpolation with anti-aliasing
Note - I also made a data set of 308x308 images not seen here

```python
# New image with new updated bounding box
old_xmin = aug_train['american_pit_bull_terrier_108']['xmin']
old_ymin = aug_train['american_pit_bull_terrier_108']['ymin']
old_xmax = aug_train['american_pit_bull_terrier_108']['xmax']
old_ymax = aug_train['american_pit_bull_terrier_108']['ymax']

img = aug_train['american_pit_bull_terrier_108']['img']
new_img = resize(img,(416,416,3))
scale_y = 416 / img.shape[0]
scale_x = 416 / img.shape[1]

xmin = old_xmin * scale_x
ymin = old_ymin * scale_y
xmax = old_xmax * scale_x
ymax = old_ymax * scale_y
width = xmax - xmin
length = ymax - ymin

fig,ax = plt.subplots(1)
ax.imshow(new_img)
rect = patches.Rectangle((xmin,ymin),width,length,linewidth=1,edgecolor='r',facecolor='none')
ax.add_patch(rect)

plt.show()
print("Old shape:",img.shape)
print("New shape:",new_img.shape)
```

```python
# Dictionary to hold resized data
# Note this dictionary will have a "y" key which represents the y data for the model which corresponds w x data: img
resized_train = {}
```

```python
"""
Resizing data and updating bounding box coordinates - all saved to a new dictionary: resized_train
"""
for animal in aug_train.keys():
    species = aug_train[animal]['species']
    old_xmin = aug_train[animal]['xmin'] # Bounding box coordinates
    old_ymin = aug_train[animal]['ymin']
    old_xmax = aug_train[animal]['xmax']
    old_ymax = aug_train[animal]['ymax']
    
    old_img = aug_train[animal]['img'] # Resizing image
    new_img = resize(old_img,(416,416,3))
    
    scale_y = 416 / old_img.shape[0] # Scaling factor for bounding box
    scale_x = 416 / old_img.shape[1]

    xmin = old_xmin * scale_x # Updating bounding box coords
    ymin = old_ymin * scale_y
    xmax = old_xmax * scale_x
    ymax = old_ymax * scale_y
    
    width = xmax - xmin # Updating bounding box length,width
    length = ymax - ymin
    
    # Adding new data to dict
    resized_train[animal] = {"species":species,"xmin":xmin,"xmax":xmax,"ymin":ymin,"ymax":ymax,"width":width,"length":length,"img":new_img,"y":0}
```

```python
# Testing my updated images
img = resized_train['american_pit_bull_terrier_129_2']['img']
width = resized_train['american_pit_bull_terrier_129_2']['width']
length = resized_train['american_pit_bull_terrier_129_2']['length']
xmin = resized_train['american_pit_bull_terrier_129_2']['xmin']
ymin = resized_train['american_pit_bull_terrier_129_2']['ymin']

fig,ax = plt.subplots(1)
ax.imshow(img)
rect = patches.Rectangle((xmin,ymin),width,length,linewidth=1,edgecolor='r',facecolor='none')
ax.add_patch(rect)

plt.show()
print("Species:",resized_train['american_pit_bull_terrier_129_2']['species'])
```

Creating the y data for each of the images. This will consist of a 13x13x14 matrix, consisting of 13x13 cells for each picture each being 32x32 pixels. The channel of 14 for each cell will consist of 2x (boolean value representing whether the bounding box is in that region, (x,y) coordinates representing the midpoint of the bounding box, (width,height) of the box, and 2 final values representing whether there is a cat or a dog being representing by the bounding box).

```python
# Showing how I am getting the midpoint of the bounding box and subsequent x and y cells the midpoint falls into
img = resized_train['american_pit_bull_terrier_129_2']['img']
width = resized_train['american_pit_bull_terrier_129_2']['width']
length = resized_train['american_pit_bull_terrier_129_2']['length']
xmin = resized_train['american_pit_bull_terrier_129_2']['xmin']
ymin = resized_train['american_pit_bull_terrier_129_2']['ymin']

cell_size = 32
mid_x = xmin + width / 2
mid_y = ymin + length / 2
cell_x = math.ceil(mid_x/cell_size) # Which cell in X direction, 1-indexed
cell_y = math.ceil(mid_y/cell_size)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x_ticks = np.arange(0, 417, cell_size) # Ticks corresponding to each cell
y_ticks = np.arange(0, 417, cell_size)

ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)
ax.grid(which='both',color='r')
ax.imshow(img)
bounding_box = patches.Rectangle((xmin,ymin),width,length,linewidth=1,edgecolor='b',facecolor='none')
midpoint = patches.Circle((mid_x,mid_y), radius=5)
ax.add_patch(bounding_box)
ax.add_patch(midpoint)

plt.show()

print("X cell:",cell_x)
print("Y cell:",cell_y)
```

```python
# Dictionary holding image data and "y" representation
final_data = {}
```

```python
# Making the 7x7x7 output matrix for each image
# Note that width,height are normalized by image height and width
# Np-arrays are row x col
cell_size = 32

for animal in resized_train.keys():
    dog = 0
    cat = 0
    
    img = resized_train[animal]['img']
    width = resized_train[animal]['width'] # Metrics per animal in dataset
    length = resized_train[animal]['length']
    xmin = resized_train[animal]['xmin']
    ymin = resized_train[animal]['ymin']
    species = resized_train[animal]['species']
    
    if species == "2": # dog
        dog = 1
    elif species == "1": # cat
        cat = 1
    
    mid_x = xmin + width / 2
    mid_y = ymin + length / 2
    cell_x = math.ceil(mid_x/cell_size) # Which cell in X direction
    cell_y = math.ceil(mid_y/cell_size)
    cell_x -= 1 # 1-indexed
    cell_y -= 1
    # Get (x,y) based on each cell being indexed from (0,0)-(1,1)
    the_mid_x = (mid_x - (cell_size*cell_x)) / cell_size
    the_mid_y = (mid_y - (cell_size*cell_y)) / cell_size
    
    width = width / 416 # Normalizing length and width
    length = length / 416
    
    y = np.zeros((13,13,7)) # Grid of y data initialized to all zeros
    y[(cell_y),(cell_x)] = np.array([1,the_mid_x,the_mid_y,width,length,dog,cat])
    y.shape = (13,13,7)
    final_data[animal] = {"X":img,"y":y}
```

```python
# Testing final images
img = final_data['american_pit_bull_terrier_129_2']['X']
width = final_data['american_pit_bull_terrier_129_2']['y'][2][5][3]
length = final_data['american_pit_bull_terrier_129_2']['y'][2][5][4]
xmin = final_data['american_pit_bull_terrier_129_2']['y'][2][5][1]
ymin = final_data['american_pit_bull_terrier_129_2']['y'][2][5][2]

awidth = width * 416
alength = length * 416
mid_x = xmin * 32 + (32*5) # Algebra to get back to actuall box coordinates from normalized coords
mid_y = ymin * 32 + (32*2)
min_x = mid_x - (awidth/2)
min_y = mid_y - (alength/2)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x_ticks = np.arange(0, 417, cell_size)
y_ticks = np.arange(0, 417, cell_size)

ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)
ax.grid(which='both',color='r')
ax.imshow(img)
bounding_box = patches.Rectangle((min_x,min_y),awidth,alength,linewidth=1,edgecolor='b',facecolor='none')
midpoint = patches.Circle((mid_x,mid_y), radius=10)
ax.add_patch(bounding_box)
ax.add_patch(midpoint)

plt.show()

# Y representation 
final_data['american_pit_bull_terrier_129_2']['y'][2][5]
```
