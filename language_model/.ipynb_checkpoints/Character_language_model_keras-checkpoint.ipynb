## Simple RNN language model to predict plausible dinosaur names

This is a similar model compared to the tensorflow implementation within this folder. This model attempts to correctly predict each and every y<sup>t+1</sup> given x<sup>t</sup> and previous lstm states. 

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import latex
from sklearn.utils import shuffle
np.random.seed(10)
from keras import backend as K
from keras.models import load_model, Model
from keras.layers import Dense, Activation, Dropout, Input, LSTM, Reshape, Lambda, RepeatVector
from keras.initializers import glorot_uniform
from keras.optimizers import Adam
from keras.backend import argmax

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Helper functions for handling data

```python
# Reading in data - creating dictionaries specifying how to encode data (into characters)
chars = []
all_names = []
with open("../../data/dino_names/names.txt") as names:
    lines = names.readlines()
    for line in lines:
        line = line.lower()
        for char in line:
            if char not in chars:
                chars.append(char)
        all_names.append(line) # \n included
                
print(chars)
print("no chars:",len(chars))
```

```python
# Creating encoder and decoder
i = 1
encoder = {"\n":0} # char to id, set \n to be 0
decoder = {0:"\n"} # id to char

for char in chars:
    if char != "\n":
        encoder[char] = i
        decoder[i] = char
        i += 1
```

```python
# returns one hot encoding for a particular character
def get_one_hot_encoding(char):
    char_enc = encoder[char]
    hot_vec = np.zeros((27,1)) # vocab_size = 27
    hot_vec[char_enc] = 1
    hot_vec = hot_vec.T # shape (1,27)
    return hot_vec
```

```python
# Returns the one-hot-encoding representation for a set of characters, i.e. "abc"
def get_character_encoding(chars):
    ret_enc = []
    m = len(chars)
    for c in chars:
        c_one_hot = get_one_hot_encoding(c) # one-hot vector
        ret_enc.append(c_one_hot)
        
    ret_enc = np.array(ret_enc)
    ret_enc.shape = (3,27) # shape (num_input, 1) for x variable
    return ret_enc
```

```python
# returns a list of ((3,27),(3,27)) pairs corresponding with ((x_train_set),(y_output_set))
# the y output is is set y^<t+1> correspond to set x^<t>
def get_training_examples(lis_names,seed):
    shuf_names = shuffle(lis_names,random_state=seed) # shuffling data
    all_names = "".join(shuf_names) # create one long string of names
    training_set = []
    i = 3
    while (i < len(all_names)):
        x_temp = get_character_encoding(all_names[i-3:i]) # x^<t>
        y_temp = get_character_encoding(all_names[i-2:i+1]) # y^<t+1>
        if x_temp[0] != "\n":
            training_set.append((x_temp,y_temp)) # only append if example doesnt start with \n
        i += 3
    
    return training_set
```

```python
# Getting initial training set / testing func.
training_set = get_training_examples(all_names,seed=0)
print(len(training_set))
print("x shape:", training_set[100][0].shape)
print("y shape:", training_set[100][1].shape)
```

```python
# Getting separate lists for X and y input to model
X = [i[0] for i in training_set]
X = np.array(X)
y = [i[1] for i in training_set]
y = np.array(y)
print(X.shape)
y.shape = (3,6636,27)
print(y.shape)
```

### Keras Model

```python
# Core structures used in the model, 128 hidden units for the lstm cell, uni-directional
reshapor = Reshape((1, 27)) # reshape to one-hot-encoding shape
LSTM_cell = LSTM(128, return_state = True) # get the hidden state of the lstm given an input
densor = Dense(27, activation='softmax') 
```

```python
# The input sequence will always be 3 given we are using 3 sets of chars to predict the next
def model():
    X = Input(shape=(3, 27)) # length of x input is (3,27) - 3 one-hot-encodings
    a0 = Input(shape=(128,), name='a0') # hidden state is 128D
    c0 = Input(shape=(128,), name='c0')
    a = a0
    c = c0
    outputs = [] # save all outputs
    
    for i in range(3): # go through input sequence
        x = Lambda(lambda x: X[:,i,:])(X) # get correct input character
        x = reshapor(x)
        a, _, c = LSTM_cell(x, initial_state=[a, c])
        out = densor(a)
        outputs.append(out)
        
    lang_model = Model(input=[X, a0, c0], output=outputs)
    return lang_model
```

```python
lang_model = model()
```

```python
opt = Adam(lr=0.01, beta_1=0.9, beta_2=0.999, decay=0.01)
lang_model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
m = 6636
a0 = np.zeros((m,128))
c0 = np.zeros((m,128))
```

```python
lang_model.fit([X, a0, c0], list(y), epochs=1000)
```

```python
def one_hot(x):
    x = K.argmax(x)
    x = tf.one_hot(x, 27) 
    x = RepeatVector(1)(x)
    return x
```
